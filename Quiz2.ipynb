```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python

data = df.drop(["date", "lights"], axis=1)


X = data.drop("Appliances", axis=1)
y = data["Appliances"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


model = LinearRegression()
model.fit(X_train_scaled, y_train)


y_pred = model.predict(X_test_scaled)


mae = mean_absolute_error(y_test, y_pred)


print(f"Mean Absolute Error: {mae:2f}")
```

```python
rss = mean_squared_error(y_test, y_pred) * len(y_test)

print(f"Residual Sum of Squares: {rss:2f}")
```

```python
rmse = mean_squared_error(y_test, y_pred, squared=False)


print(f"Root Mean Squared Error: {rmse:.2f}")
```

```python
y_pred = model.predict(X_test_scaled)


r2 = r2_score(y_test, y_pred)


print(f"Coefficient of Determination (R-squared): {r2:.2f}")
```

```python
feature_weights = pd.DataFrame({'Feature': X.columns, 'Weight': model.coef_})

feature_weights = feature_weights.sort_values(by='Weight')


feature_lowest = feature_weights.iloc[0]['Feature']
feature_highest = feature_weights.iloc[-1]['Feature']

# Print the features with the lowest and highest weights
print("Feature with the lowest weight:", feature_lowest)
print("Feature with the highest weight:", feature_highest)
```

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge(alpha=0.4)
model.fit(X_train_scaled, y_train)


y_pred = model.predict(X_test_scaled)


baseline_rmse = mean_squared_error(y_test, y_pred, squared=False)


print("RMSE for the baseline model:", baseline_rmse)
```

```python
from sklearn.linear_model import Lasso
```

```python
model = Lasso(alpha=0.001)
model.fit(X_train_scaled, y_train)


feature_weights = pd.DataFrame({'Feature': X.columns, 'Weight': model.coef_})


non_zero_features = feature_weights[feature_weights['Weight'] != 0].shape[0]


print("Number of features with non-zero weights:", non_zero_features)
```
